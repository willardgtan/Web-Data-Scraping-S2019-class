```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
### This is loading requests, beautifulsoup and pandas. 
```

xmas_bo_raw = requests.get('https://www.the-numbers.com/box-office-chart/daily/2018/12/25').text
### here he combined the request get with the .text so it's in one move instead of doing another r=r.text or somehting similar

```python
xmas_bo_soup = BeautifulSoup(xmas_bo_raw, 'html.parser')
### this is the christmas box office soup variable..whose value is coming from the requests.get.txt and the html.parser
```

```python
xmas_bo_tables = xmas_bo_soup.find_all('table')
### this is a new variable tables which finds all tables
```

```python
print(xmas_bo_tables[1])
### this is showing the table indexed at 1.  there are two tables, one indexed at zero, one indexed at one
```

```python
xmas_bo_trs = xmas_bo_tables[1].find_all('tr')

### this is a new variable trs which finds all of the 'tr' within the table indexed at 1.  
###  th  is table headers
### Tr is table row
### td is table data
```

```python
xmas_bo_trs[0]
### this is the list of the table row at index of zero, or basically the header.  All are labeld with the tr so its a table 
### header.
```

```python
xmas_bo_trs[1].text.split('\n')
### this shows us the characters between the tables, which are commas
```

```python
cleaned_xmas_bo_rows =[]
### create new dictionary

for row in xmas_bo_trs[1:]:
    ###loop through all non header rows (by starting at row 1)
    cleaned_row = row.text.split('\n')
    ### this is splitting the lines into new lines
    cleaned_xmas_bo_rows.append(cleaned_row)

cleaned_xmas_bo_rows[:2]
```

```python
xmas_bo_df = pd.DataFrame(cleaned_xmas_bo_rows)


xmas_bo_df = xmas_bo_df.drop(columns=[0,12])
xmas_bo_df.columns = ['Rank','Last rank','Movie','Distributor','Gross',
                      'Change','Lst Week','Theaters','Per theater','Total gross',
                      'Days',]

xmas_bo_df.head()
```
